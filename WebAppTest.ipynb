```python
#import library
import os
import cv2
import random
import warnings

import numpy as np
import pandas as pd
import seaborn as sns
import streamlit as st
import tensorflow as tf
import random as python_random
import matplotlib.pyplot as plt

from tensorflow import keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
WIDE = 128
BATCH_SIZE = 1
ROUND = 100
```

```python
def load_ben_color(img):
    image = cv2.addWeighted(img,2, cv2.GaussianBlur( img , (0,0) , WIDE/5) ,-2 ,100)
    return image
```

```python
model = keras.models.load_model('Prepro94.75')
```

```python
correct = 0
wrong = 0
for i in range(ROUND):
    path = r'D:\ML\DRDATASET\Dataset2\Unclean\2Class\RGB'
    files = os.listdir(path)
    randomFile = random.choice(files)
    files2 = os.listdir(path+'/'+randomFile)
    randomFile2 = random.choice(files2)
    if randomFile == 'DR0':
        Classnum = 0
    else :
        Classnum = 1
    print('Class      : DR', Classnum)
    print('Image name :', randomFile2)
    img_bgr = cv2.imread(path+'/'+randomFile+'/'+randomFile2)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_rgb_resized = cv2.resize(img_rgb,(WIDE,WIDE))
    img_rgb_resized  = img_rgb_resized.reshape(BATCH_SIZE,WIDE,WIDE,3)

    datagen = ImageDataGenerator(rescale= 1./255, preprocessing_function=load_ben_color)
    testdata = datagen.flow(img_rgb_resized)

    predict = model.predict(testdata,verbose=0)
    Classes = np.argmax(predict)
    print('Prediction for this image : DR',Classes)
    if Classnum == Classes :
        print('Prediction is Correct!!')
        correct += 1
    else : 
        print('Prediction is Wrong!!')
        wrong += 1
    print()
print('Summary : Correct',correct,'Wrong',wrong)
print('          Accuracy',(correct/ROUND)*100,'%')
```
